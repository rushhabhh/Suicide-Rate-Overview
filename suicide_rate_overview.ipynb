```python
import pandas as pd
import numpy as np


import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('/Users/rush/Documents/Documents - Rushabhâ€™s MacBook Air/Data/Suicide Rate Overview/suicide_overview.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'gdp_for_year($) ': 'gdp_for_year'}, inplace=True)
df.rename(columns={'gdp_per_capita($)': 'gdp_per_capita'}, inplace=True)
df.rename(columns={'suicides/100k pop': 'suicides/100k'}, inplace=True)
df.rename(columns={'HDI for year': 'hdi_for_year'}, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df['country'] = df['country'].astype('category')
df['sex'] = df['sex'].astype('category')
df['age'] = df['age'].astype('category')
df['country-year'] = df['country-year'].astype('category')
df['generation'] = df['generation'].astype('category')

df.dtypes
```

```python
df.drop(columns=['country-year'], inplace=True)
df.head()
```

```python
df['country'].unique()
```

```python
df['country'] = df['country'].replace('United States of America', 'United States')
df['country'] = df['country'].replace('Russian Federation', 'Russia')

df['country'].unique()
```

```python
df['generation'].unique()
```

```python
df['generation'] = df['generation'].replace('Millenials', 'Generation Y')
df['generation'].unique()
```
